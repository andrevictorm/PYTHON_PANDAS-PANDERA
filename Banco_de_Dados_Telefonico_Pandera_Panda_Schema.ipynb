```python
import pandas as pd
df = pd.read_csv("https://storage.googleapis.com/bucket-dataprothiago/Acessos_Banda_Larga_Fixa_2021.csv", sep=";")
```

```python
import pandera as pa
```

```python
df.loc[df.Tecnologia == "WIMAX" , ['Meio de Acesso']] = 'Ar'
```

```python
df.rename(columns={'Tipo de Pessoa': 'Tipo_de_Pessoa'}, inplace=True)
```

```python
sorted(pd.unique(df['Tecnologia']))
```

```python
df.loc[df.Tecnologia == "OFDMA/TDD" , ['Tecnologia']] = '3G'
df.loc[df.Tecnologia == "EDGE" , ['Tecnologia']] = '4G'
df.loc[df.Tecnologia == "LTE" , ['Tecnologia']] = '2G'
df.loc[df.Tecnologia == "GSM" , ['Tecnologia']] = '2G'
df.loc[df.Tecnologia == "OFDMA/TDD" , ['Tecnologia']] = '3G'
df.loc[df.Tecnologia == "FTTH" , ['Tecnologia']] = 'FIBRA'
df.loc[df.Tecnologia == "FTTB" , ['Tecnologia']] = 'FIBRA'
```

```python
df.replace(['Indeterminado'],pd.NA, inplace=True)
```

```python
schema = pa.DataFrameSchema(
    columns={
       'codigo_ocorrencia':pa.Column(pa.Int),
       'codigo_ocorrencia2':pa.Column(pa.Int),
       'ocorrencia_classificacao':pa.Column(pa.String),
       'ocorrencia_cidade':pa.Column(pa.String),
       'ocorrencia_uf':pa.Column(pa.String,pa.Check.str_length(2,2),nullable=True),
       'ocorrencia_aerodromo':pa.Column(pa.String,pa.Check.str_length(4,4),nullable=True),
       'ocorrencia_dia':pa.Column(pa.DateTime),
       'ocorrencia_hora':pa.Column(pa.String,pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9]):([0-5][0-9])?$'),nullable=True),
       'total_recomendacoes':pa.Column(pa.Int),
    }     
)
```
